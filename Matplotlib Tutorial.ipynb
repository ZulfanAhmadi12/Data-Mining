### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch, When saving probably should use 300 if possible)
plt.figure(figsize=(5,3), dpi=100)

# Line 1
#Keyword argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, linestyle='--',marker='.', markersize=10, markeredgecolor='black')

# Use shorthand notation to costume your plot faster
# fmt = '[color][marker][line]'
plt.plot(x,y,'b^--', label='2*x')

## Line Number Two
# you can make the same line to be different, look at the red line
# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)
# Plot part of the graph as line
plt.plot(x2[:6],x2[:6]**2,'r^-', label='X^2')
#Plot remainder of graph as a dot
plt.plot(x2[5:],x2[5:]**2, 'r.--')

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!',fontdict={'fontsize':15, 'fontname': 'Comic Sans MS'})

# X and Y Labels
plt.xlabel('X Axis!',fontdict={'fontsize':13, 'fontname': 'Comic Sans MS'})
plt.ylabel('Y Axis!',fontdict={'fontsize':13, 'fontname': 'Comic Sans MS'})

#This will specify your X and Y manually/ X, Y axis Tickmarks (scale of your graph) 
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

### Load Fifa Data

```python
import pandas as pd
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

### Histograms

```python
bins =[40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins)

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skils in FIFA 2018')



plt.show()
```

### Pie Charts

```python
left = fifa.loc[fifa['Preferred Foot']== 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']== 'Right'].count()[0]

labels = ['Left Foot','Right Foot']
colors = ['#abcdef','#aabbcc']

plt.pie([left,right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
#this is how i change the data type of Weight column which was originally a string into an integer
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight <150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >=175) & (fifa.Weight <200)].count()[0]
heavy = fifa[(fifa.Weight >=200)].count()[0]


weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125','125-150','150-175','175-200','Over 200']
explode = (.4,.2,.0,.0,.4)

plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.7, explode=explode)
plt.show()
```

```python
#comparing 2 teams with box plot using data from their Overall score column
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club =='Real Madrid']['Overall']
mu = fifa.loc[fifa.Club =='Manchester United']['Overall']

labels = ['FC Barcelona',' Real Madrid', 'Manchester United']

#if you want to change the color with the same color you can use it in the boxplot
boxes = plt.boxplot([barcelona,madrid,mu], labels=labels, patch_artist=True,medianprops={'linewidth':2})

#use this if you want to change the color of box each differently
for box in boxes['boxes']:
    #Set edge color and width
    box.set(color='#4286f4', linewidth=2)
    # Change fill color
    box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```
