```python
import pandas as pd
import numpy as np

data = pd.read_csv('https://raw.githubusercontent.com/realpython/python-data-cleaning/master/Datasets/BL-Flickr-Images-Book.csv')
data
```

```python
data.info()
```

```python
# inplace berarti apa yang di drop itu di update, axis 1 = column vertical, axis 0 = rows horizontal
drop = ['Edition Statement',
           'Corporate Author',
           'Corporate Contributors',
           'Former owner',
           'Engraver',
           'Contributors',
           'Issuance type',
           'Shelfmarks']
data.drop(drop, inplace = True, axis = 1)
data.head()
```

```python
data.info()
```

```python
data['Identifier'].is_unique
```

```python
data = data.set_index('Identifier')
data.head()
```

```python
extr = data['Date of Publication'].str.extract(r'^(\d{4})',expand=False)
extr.head(40)
```

```python
data['Date of Publication'] = pd.to_numeric(extr)
data['Date of Publication'].dtype
```

```python
pub = data['Place of Publication']
london = pub.str.contains('London')
london[:5]
```

```python
oxford = pub.str.contains('Oxford')
oxford[:5]
```

```python
data['Place of Publication']=np.where(london,'London',np.where(oxford,'Oxford',pub.str.replace('-', ' ')))
data['Place of Publication'].head(70)
```
