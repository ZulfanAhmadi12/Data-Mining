```python
weather = pd.read_csv('data/weather.csv')
```

```python
#when you see the values of a rows and column is almost the same with many duplicate like tmax and tmin
#it means the data isnt Tidy, they have variable in rows and columns
weather
```

```python
#you cant use weather.melt not pd.melt, it call the exact same code as pd, in pivot table you need to use dataframe(name) dot pivot_table not pandas name as pd

weather_melt = weather.melt(
    weather,
    id_vars=['id','year','month','element'],
    var_name='day',
    value_name='temp'
)
```

```python
weather_melt
```

```python
#the difference between pivot and pivot_table, for pivot you are assuming there are no duplicate values, for pivot table its the opposite
#index parameters use to specify which column you dont want to move 
#columns parameter use to pivot this element columns it means to look at this element columns for each unique value in this column, its going to create a totally separate column

weather_tidy = weather_melt.pivot_table(
    index=['id','year','month','day'],
    columns='element',
    values='temp'
)
```

```python
#if you want a regular flat dataframe, sometimes you would have to run reset_index twice
weather_tidy.reset_index()
```

```python
#how to filtering your dataset with drop duplicate, when you have many duplicate values in a column

billboard_songs = billboard_melt[['year','artist','track','time']]

```

```python
billboard_songs.head()
```

```python
#this is how you drop the duplicated values
billboard_songs = billboard_songs.drop_duplicates()
```

```python
billboard_songs
```

```python
#how to join the tables, you need a key, so this is how to generate a key
#range is a function in python uses as generator, you give it a number and it'll generate a value, think of it as a list from zero to that number
billboard_songs['id'] = range(len(billboard_songs))
```

```python
billboard_songs.head()
```

```python
#index=false mean you dont need index column 0,1,2,3,4... in your dataset csv
billboard_songs.to_csv('billboard_songs.csv', index=False)
```

```python
#combine the songs table and the original one, we combine songs table because we need the id column we just created
#when you press shift+tab, you'll see left will be billboard_melt, right is billboard_songs,how by defualt is inner meaning that only the tables that found matches will be returned back
#if the column names of both left and right df is the same you can say on=None, but if the column names of both sides different you can say left_on,right_on
# we use on here because we have the same column in both sides
billboard_ratings = billboard_melt.merge(
    billboard_songs, on=['year','artist','track','time']

)
```

```python
billboard_ratings.head(100)
```

```python
billboard_ratings = billboard_ratings[['id','date.entered','Week','Rating']]
```

```python
billboard_ratings.to_csv('billboard_ratings.csv', index=False)
```
